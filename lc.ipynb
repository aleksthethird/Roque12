```python
from astropy.io import fits as pyfits
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
dat = pyfits.getdata('r12.fits')
jd = dat.field(1)
jd = jd - 290.0
phot1 = dat.field(0)
phot2 = dat.field(2)
plt.plot(jd[0:131],phot2[0:131],'bo')
print(len(phot2))
print(np.std(phot2[0:131]))
```

```python
jdat = np.loadtxt('jdata.txt')
j_time = jdat[:,0] - 306.0 + 16.0
j_mag = jdat[:,2]
plt.plot(j_time,j_mag)
```

```python
plt.figure(figsize=(8, 6))
plt.rc('font', family='serif',size=12)
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
plt.xlim(0,18)
plt.ylim(-0.75,0.2)
plt.xlabel('JD - 2452550')
plt.ylabel('Relative magnitude')
plt.plot(jd,phot2,'bo',label="I-band")
plt.title('Roque 12 lightcurve, 2-20 Oct 2002')
```

```python
plt.figure(figsize=(8, 6))
plt.rc('font', family='serif',size=12)
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
plt.xlim(16.4,16.75)
plt.ylim(-0.75,0.2)
plt.xlabel('JD - 2452550')
plt.ylabel('Relative magnitude')
plt.plot(jd,phot2,'bo',label="I-band")
plt.plot(j_time,j_mag,'ro',label="J-band")
plt.title('Roque 12 lightcurve, 18-19 Oct 2002')
plt.legend(loc=3)
# add error: 0.07 in J-band
```
