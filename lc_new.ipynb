```python
from astropy.io import fits as pyfits
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
from IPython.display import set_matplotlib_formats
set_matplotlib_formats('png', 'pdf')
```

```python
dat = pyfits.getdata('r12.fits')
jd = dat.field(1)
jd = jd - 290.0
phot1 = dat.field(0)
phot2 = dat.field(2)
plt.plot(jd[0:131],phot2[0:131],'bo')
print(len(phot2))
print(np.std(phot2[0:131]))
print(phot2[130:140],jd[130:140])
```

```python
jdat = np.loadtxt('jdata.txt')
j_time = jdat[:,0] - 306.0 + 16.0
j_mag = jdat[:,2]
plt.plot(j_time,j_mag)
j_err = np.array([0.07,0.07,0.07,0.07,0.07,0.07,0.07,0.07,0.07])
```

```python
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='15')
plt.rc('ytick', labelsize='15')
fig = plt.figure(figsize=(10, 8))
plt.xlim(0,18)
plt.ylim(-0.75,0.2)
plt.xlabel('JD - 2452550',fontsize=20)
plt.ylabel('Relative magnitude',fontsize=20)
plt.plot(jd,phot2,'bo')
plt.title('Roque 12 lightcurve, 2-20 Oct 2002',fontsize=20)
```

```python
plt.rc('font', family='serif')
plt.rc('xtick', labelsize='15')
plt.rc('ytick', labelsize='15')
fig = plt.figure(figsize=(10, 8))
plt.xlim(16.4,16.75)
plt.ylim(-0.75,0.2)
plt.xlabel('JD - 2452550',fontsize=20)
plt.ylabel('Relative magnitude',fontsize=20)
plt.plot(jd,phot2,'bo',label="I-band")
plt.plot(j_time,j_mag,'ro',label="J-band")
plt.title('Roque 12 lightcurve, 18-19 Oct 2002',fontsize=20)
plt.legend(loc=3,fontsize=15)
plt.errorbar(j_time,j_mag,j_err,color='red',ls='')
plt.plot([16.71560669,16.71560669],[-0.033-0.034,-0.033+0.034],'b')
```
